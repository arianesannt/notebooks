# Desafio estatística e álgebra linear com python - Fase 1

Esse desafio foi proposto como atividade extracurricular da PósTech Fiap Data Analytics - disciplina: Algebra linear e Estatística, Fase 1.

O objetivo é encontrar um polinômio cúbico através do método dos mínimos quadrados (MMQ) que passa pelos pontos: (-1,-10), (0,5), (1,6), (2,11) e (3,42).

Deve ser plotado um gráfico que melhor se ajusta aos pontos dados.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# pontos de dados
pontos = [(-1, -10), (0,5), (1,6), (2,11), (3,42)]

# separando x e y
x = []
y = []

for elemento in pontos:
    primeiro_elemento, segundo_elemento = elemento
    x.append(primeiro_elemento)
    y.append(segundo_elemento)



print(x)
print(y)
```

```python
# transformando as listas x e y em um array do numpy
# isso será importante para poder a aplicar o método vstack do numpy
x = np.array(x)
y = np.array(y)
```

```python
# aplicando o MMQ para encontrar os coeficientes do polinômio cúbico
A = np.vstack([x**3, x**2, x, np.ones(len(x))]).T
coeficientes, residuais, _, _ = np.linalg.lstsq(A, y, rcond=None)
```

```python
# coeficientes do polinômio cúbico
a, b, c, d = coeficientes

# gerando pontos para o polinômio cúbico
x_ajuste = np.linspace(-1, 3, 400)
y_ajuste = a * x_ajuste**3 + b * x_ajuste**2 + c * x_ajuste + d
```

```python
# plotando os pontos de dados e o polinômio cúbico ajustado
plt.figure(figsize=(5,6), layout='constrained')
plt.plot(x_ajuste, y_ajuste, label=f'Polinômio Cúbico: {a:.2f}x^3 + {b:.2f}x^2 + {c:.2f}x + {d:.2f}', color='yellow')
plt.scatter(x, y, label='Pontos de Dados', color='gray')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower right', fontsize='small')
plt.grid(True, linewidth=0.2)
plt.title('Ajuste de Polinômio Cúbico com MMQ')
plt.show()
```
